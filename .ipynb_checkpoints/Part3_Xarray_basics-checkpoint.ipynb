## 3.1.1 DataArray

1. data: One data variable (an array of data) 
2. dims: Dimension names
2. coords: Coordinate variables
4. attrs: Attribute data (metadata describing the data (optional))     

```python
np.random.seed(0)

# Define data variable (lat=6, lon=6, time=7)
temp = 15 + 8 * np.random.randn(6, 6, 7)

# Define coordinate variables
lat = [42.0, 42.5, 43, 43.5, 44, 44.5]
lon = [-99.0, -99.5, -100, -100.5, -101, -102]
time = pd.date_range('2024-06-06', periods=7)
```

```python
# Build DataArray
da = xr.DataArray(
    data = temp,
    dims = ['lon', 'lat', 'time'],
    coords=dict(lat=lat, lon=lon, time=time),
    attrs=dict(description='Ambient temperature', unit='degC')
)
```

```python
# Assign CRS
crs = 'EPSG:4326'
da = da.rio.write_crs(crs)
```

## 3.1.2 Dataset

1. data: Data variables
2. dims: Dimension names
2. coords: Coordinate variables
4. attrs: Attribute data (metadata describing the data (optional))     

```python
np.random.seed(0)

# Define additional data variable
precip = 10 * np.random.rand(6, 6, 7)
```

```python
# Build Dataset
ds = xr.Dataset(
    dict(temp=(['lon', 'lat', 'time'], temp),
        precip=(['lon', 'lat', 'time'], precip)),
    coords=dict(lat=lat, lon=lon, time=time),
    attrs=dict(description='Meteorological data', 
               source='AIBYNETO')
)
```

```python
# Add global attribute
ds.attrs['temp'] = 'Ambient temperature (degC)'
```

```python
# Add local attribute
ds['temp'].attrs['temp'] = 'Ambient temperature (degC)'
```

## 3.3 Clip data

### 3.3.1 Clipping by mask extent

```python
xmin, xmax, ymin, ymax = 100, 101, 13.4, 14.4
pop_clip = pop.sel(x=slice(xmin, xmax), y=slice(ymax, ymin))
```

```python
ax = pop_clip['band_data'].plot()
ax.axes.set_aspect('equal')
```

### 3.3.2 Clipping by shapefile

```python
bg = gpd.read_file(data_path + 'asia-administrative-boundaries/cntry_3m.shp')
```

```python
# Assign CRS to Elevation data
crs = 'EPSG:4326'
ele.rio.write_crs(crs, inplace=True)

# Change shapefile crs
bg.to_crs(crs, inplace=True)
```

```python
# Reduce resolution of Elevation data
ele = ele.coarsen(x=10, y=10, boundary='trim').mean()
```

```python
ax = ele['band_data'].plot()
ax.axes.set_aspect('equal')
```

```python
# Clip elevation data by polygons
clip_bounds = bg[bg['CNTRY_NAME'] == 'Thailand']
ele_clip = ele['band_data'].rio.clip(clip_bounds['geometry'])
```

```python
ax = ele_clip.plot()
ax.axes.set_aspect('equal')
```
